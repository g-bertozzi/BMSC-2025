getScalarData()

Returns data from a specific location and device category

Parameters: 
    Required
    - locationCode: str - ONC location code (e.g., "CF341")
    - deviceCategoryCode: str - Comma-separated device types to fetch from
        ex. CTD, OXYSENSOR, radiometer, FLNTU
    - dateFrom: str - Start date in ISO 8601 format (e.g., "2023-07-11T17:00:00.000Z")
    - dateTo: str - End date in ISO 8601 format (e.g., "2023-07-11T22:30:00.000Z")

    Optional
    - sensorCategoryCode
    - propertyCode
    - resample
    - type
    - clean

Return: a JSON fromatted block of text containing "citations", "messages", "metadata", "parameters", "next", "sensorData"

    Within the "sensorData" section we have one or more "data" sections which further contain fields including: "sampleTimes" and "values". For plotting these are the two sections of information we most need and will extract data from.

ex. Requesting data from a device that has a single sensor.

    result = getScalarData(parameters)

    exampleSensor = result["sensorData"]["data"]
    exampleTimeStamps = exampleSensor
    exampleValues = exampleSensor




Schema: 

{
  "citations": [],
  "messages": [],
  "metadata": {},
  "parameters": {},
  "next": {},
  "sensorData": [
    {
      "data": {
        ...
        "sampleTimes": [],
        "values": [],
        ...
      },
    }
  ],
  "queryUrl": "string"
}
